## Question 2

Explain the relationship between **Primary Keys** and **Foreign Keys**.

Type your answer below this line:
_______________________________________________________________________________________________________________________________
Primary Keys are the unique identifiers of the entries in a table. Foreign Keys are any column in that table used to pair up with a Primary Key of another table.

An example might be: a Students table with id numbers (primary key) for each student. Their 'homeroom_teacher_id' column can be a foreign key to the Teachers table's 'teacher_id', which is its Primary Key.



## Question 3

Explain the different types of relationships entities can have in a SQL database. 

Type your answer below this line:
_______________________________________________________________________________________________________________________________
3 types:
- One-to-one: Direct relationship of one entry to exactly one other entry in another table.
- One-to-many: Relationship of a larger-scale table ('offices') to a smaller-scale table with lots of entries for each entry in the other table ('employees')
- Many-to-many: Complex relationship between lots of entries ('employees') and lots of entries in another table ('customers'), where many customers are associated with many employees in multiple ways.


## Question 4

Explain the various types of JOINs possible with SQL. 

Type your answer below this line:
_______________________________________________________________________________________________________________________________
- Thinking of it as a Venn diagram, joins are differentiated by which tables and what intersections are included in the query.
- JOIN / LEFT JOIN: All the 'left' table entries, with any 'right' table entries that match.
- RIGHT JOIN: The reverse of LEFT joins
- INNER JOIN: All entries existing in both tables, but none that are exclusive to either
- OUTER JOIN: All entries that match only one table, but none that match both
- FULL JOIN: rarely used, matches every entry in the left and every in the right.
